```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from branca.colormap import LinearColormap
```

```python


# Sample data (replace this with actual data)
data = {
    'City': ['Paris', 'Bangkok', 'London', 'Dubai', 'Singapore',
             'New York', 'Kuala Lumpur', 'Tokyo', 'Istanbul', 'Seoul',
             'Hong Kong', 'Mumbai', 'Rome', 'Osaka', 'Taipei',
             'Mecca', 'Pattaya', 'Phuket', 'Guangzhou', 'Miami',
             'Barcelona', 'Moscow', 'Prague', 'Amsterdam', 'Vienna'],
    'Rank': list(range(1, 26)),
    'Visitors': [85.0, 76.3, 72.0, 68.0, 66.7,
                  65.2, 63.5, 62.3, 60.1, 59.3,
                  58.7, 57.0, 56.4, 55.8, 54.2,
                  53.9, 52.8, 51.7, 50.6, 49.5,
                  48.4, 47.3, 46.2, 45.1, 44.0],  # In millions
    'Latitude': [48.8566, 13.7563, 51.5074, 25.2769, 1.3521,
                 40.7128, 3.1390, 35.6762, 41.0082, 37.5665,
                 22.3193, 19.0760, 41.9028, 34.6937, 25.0330,
                 21.3891, 12.9236, 7.9519, 23.1291, 25.7617,
                 41.3851, 55.7558, 50.0755, 52.3676, 48.2082],
    'Longitude': [2.3522, 100.5018, -0.1278, 55.2962, 103.8198,
                  -74.0060, 101.6869, 139.6503, 28.9784, 126.9780,
                  114.1694, 72.8777, 12.4964, 135.5023, 121.5654,
                  39.8579, 100.8824, 98.3381, 113.2644, -80.1918,
                  2.1734, 37.6173, 14.4378, 4.9041, 16.3738]
}

df = pd.DataFrame(data)

```

```python



# Create color scale
colormap = LinearColormap(
    colors=['green', 'yellow', 'red'],
    vmin=df['Visitors'].min(),
    vmax=df['Visitors'].max()
)







```

```python
# Create base map
m = folium.Map(location=[20, 0], zoom_start=2, tiles='cartodbpositron')

# Add marker cluster
marker_cluster = MarkerCluster().add_to(m)
```

```python
# Add points to the map
for idx, row in df.iterrows():
    popup_text = f"""
    City: {row['City']}<br>
    Rank: {row['Rank']}<br>
    Visitors: {row['Visitors']}M
    """
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=10,
        popup=popup_text,
        color=colormap(row['Visitors']),
        fill=True,
        fill_color=colormap(row['Visitors']),
        fill_opacity=0.7,
        weight=1
    ).add_to(marker_cluster)
```

```python
# Add color scale legend
colormap.caption = 'Number of Visitors (Millions)'
colormap.add_to(m)

```
